````python
#1

from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test

import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5):
 print(y_test[i],'\n') # Print the label
 print(X_test[i],'\n') # Print the matrix of values
 plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
 plt.show()
````

````python
#2

print(X_train.shape)
print(X_test.shape)
print(X_train[0].shape)
print(X_test[0].shape)
print(y_train.shape)
print(y_test.shape)
````

In the training set there are 60,000 matrices of a size of 28 by 28.

In the test set there are 10,000 matrices of a size of 28 by 28.

Moreover, in the y_train there are 60,000 numeral assignments. In comparison to the y_test which has 10,000 numeral assignments. Overall the X_train[i] is a matrix value in a 2D space that is linked with the y_train[i].

````python
#3

import pandas as pd
import numpy as np

reload = 0

if reload == 1:
  Z_train = []
  for i in range(60000):
    row = X_train[i].reshape((1, 784))
    Z_train.append(row[0])
  Z_train = pd.DataFrame(Z_train)
  Z_train.to_csv('Z_train.csv')

  Z_test = []
  for i in range(len(y_test)):
    row = X_test[i].reshape((1, 784))
    Z_test.append(row[0])
  Z_test = pd.DataFrame(Z_test)
  Z_test.to_csv('Z_test.csv')
else:
  Z_train = pd.read_csv('/content/Z_train.csv')
  Z_test = pd.read_csv('/content/Z_test.csv')
````

````python
#4

from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

mnist = fetch_openml('mnist_784', version=1)
Z = mnist.data
y = mnist.target.astype(int)

print("Z shape:", Z.shape)
print("y shape:", y.shape)

Z_train, Z_test, y_train, y_test = train_test_split(Z, y, test_size=0.2, random_state=42)

k_bar = 50
k_grid = np.arange(2, k_bar)
accuracy = np.zeros(k_bar)

for k in range(k_bar):
    knn = KNeighborsClassifier(n_neighbors = k + 1)
    predictor = knn.fit(Z_train, y_train)
    accuracy[k] = knn.score(Z_test, y_test)

accuracy_max = np.max(accuracy)
max_index = np.where(accuracy==accuracy_max)
k_star = k_grid[max_index]
print(k_star)

plt.plot(np.arange(0, k_bar), accuracy)
plt.xlabel("k")
plt.title("optimal k:"+str(k_star))
plt.ylabel('Accuracy')
plt.show()
````

````python
#5

knn = KNeighborsClassifier(n_neighbors = 1)
predictor = knn.fit(Z_train.values, y_train)
y_hat = predictor.predict(Z_test.values)

accuracy = knn.score(Z_test.values, y_test)
print('Accuracy: ', accuracy)

pd.crosstab(y_test, y_hat)
````
